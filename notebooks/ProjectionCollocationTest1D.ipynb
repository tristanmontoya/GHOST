```python
# GHOST - Projection Test (1D)

import sys  
import numpy as np
import modepy as mp
sys.path.insert(0, '../src')
from Mesh import Mesh1D
from Discretization import SpatialDiscretization
from Solver import Solver

# set up mesh
mesh = Mesh1D("mesh1d", 0.0, 1.0, 4)

# discretization order
p = 2

# volume collocation degree
p_omega = p

# solver parameters
params = {"project_title": "proj_test_1d",
        "problem": "projection",
         "initial_condition": "sine",
         "wavelength": np.array([1.0]),
         "integration_type": "collocation",
         "solution_degree": p,
         "volume_collocation_degree": p_omega}

# set up solver
projection_solver = Solver(params,mesh)
```

```python
projection_solver.load_solution()
projection_solver.post_process()
projection_solver.plot(plot_numerical=True, 
                       plot_exact=True, 
                       plot_nodes=True)
```
