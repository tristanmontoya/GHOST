```python
# GHOST - Projection Test (1D)

import sys  
import numpy as np
import modepy as mp
sys.path.insert(0, '../src')
from Mesh import Mesh1D
from Discretization import SpatialDiscretization
from Solver import Solver

# set up mesh
mesh = Mesh1D("test", 0.0, 1.0, 10)
mesh.make_periodic([1,2])

# discretization order
p = 3

# volume quadrature order
tau = 2*p+5
mu = None

# solver parameters
params = {"project_title": "proj_test_1d",
        "problem": "projection",
         "initial_condition": "sine",
         "wavelength": np.array([0.25]),
         "integration_type": "quadrature",
         "solution_degree": p,
         "volume_quadrature_degree": tau,
         "facet_quadrature_degree": mu}

# set up solver
projection_solver = Solver(params,mesh)
projection_solver.plot_exact_solution()
```

```python
projection_solver.run()
projection_solver.plot_numerical_solution()
```
